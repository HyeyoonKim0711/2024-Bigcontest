```python
import pandas as pd
```

```python
resid_population = pd.read_csv('지역별(행정동) 성별 연령별 주민등록 인구수_20231031.csv', encoding='cp949')
resid_population.head()
```

```python
def resid_population_group(data, code):
  data = resid_population[resid_population['행정기관코드'] == code]

  age_under_10 = data.iloc[0,8:28].sum() + data.iloc[0,119:139].sum()
  age_20_30 = data.iloc[0,28:48].sum() + data.iloc[0,139:159].sum()
  age_40_50 = data.iloc[0,48:68].sum() + data.iloc[0,159:179].sum()
  age_above_60 = data.iloc[0,68:119].sum() + data.iloc[0,179:230].sum()

  total_list = [age_under_10, age_20_30, age_40_50, age_above_60]

  return total_list
```

```python
codes = [4817056500, 5113051500, 4122060000, 4513060500, 4111574000, 4313051500, 4713055000,
         4413151000, 1141055500, 4719025000, 4511171100, 2771026500, 4613057000, 5111054500,
         4728057000, 4717058500, 4471025000, 5177025000, 2723063000, 4817056500, 1123054500,
         5123051000, 5177025600, 4572025000, 4571025000, 4825051000, 3114052000, 5011053000,
         4521051000, 2711057500, 4127155000, 5011051000, 4888025000, 4715053600, 4127158000,
         4514052000, 2611057000, 2623052000, 2811053000, 3011051500]
```

```python
group_list = []

for code in codes:
  group_list.append(resid_population_group(resid_population, code))
```

```python
final_data = pd.DataFrame(group_list, columns = ['age_under_10', 'age_20_30', 'age_40_50', 'age_above_60'])
final_data
```

```python
final_data.to_csv('주민등록인구.csv', index=False)
```
