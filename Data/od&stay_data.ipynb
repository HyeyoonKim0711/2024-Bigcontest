### 유동인구 데이터

```python
def od_population(code):
    zip_file_path = 'C:\\Users\\pjh08\\OneDrive\\바탕 화면\\데이터분석 분야_로우데이터.zip'

    df_list = []

    # ZIP 파일 내의 특정 파일을 직접 읽기
    with zipfile.ZipFile(zip_file_path, 'r') as outer_zip:
        # 내부 ZIP 파일을 메모리 상에서 읽기
        with outer_zip.open('od_20230901_10.zip') as inner_zip_file:
            # 내부 ZIP 파일을 다시 ZipFile 객체로 처리
            with zipfile.ZipFile(BytesIO(inner_zip_file.read())) as inner_zip:
                # CSV 파일을 메모리 상에서 바로 읽기
                for csv_name in inner_zip.namelist():
                    with inner_zip.open(csv_name) as csv_file:
                        df = pd.read_csv(csv_file)
                        df = df[df['dest_hdong_cd'] == code]
                        df_list.append(df)

    with zipfile.ZipFile(zip_file_path, 'r') as outer_zip:
        # 내부 ZIP 파일을 메모리 상에서 읽기
        with outer_zip.open('od_20230911_20.zip') as inner_zip_file:
            # 내부 ZIP 파일을 다시 ZipFile 객체로 처리
            with zipfile.ZipFile(BytesIO(inner_zip_file.read())) as inner_zip:
                # CSV 파일을 메모리 상에서 바로 읽기
                for csv_name in inner_zip.namelist():
                    with inner_zip.open(csv_name) as csv_file:
                        df = pd.read_csv(csv_file)
                        df = df[df['dest_hdong_cd'] == code]
                        df_list.append(df)

    with zipfile.ZipFile(zip_file_path, 'r') as outer_zip:
        # 내부 ZIP 파일을 메모리 상에서 읽기
        with outer_zip.open('od_20230921_30.zip') as inner_zip_file:
            # 내부 ZIP 파일을 다시 ZipFile 객체로 처리
            with zipfile.ZipFile(BytesIO(inner_zip_file.read())) as inner_zip:
                # CSV 파일을 메모리 상에서 바로 읽기
                for csv_name in inner_zip.namelist():
                    with inner_zip.open(csv_name) as csv_file:
                        df = pd.read_csv(csv_file)
                        df = df[df['dest_hdong_cd'] == code]
                        df_list.append(df)

    with zipfile.ZipFile(zip_file_path, 'r') as outer_zip:
        # 내부 ZIP 파일을 메모리 상에서 읽기
        with outer_zip.open('od_20231001_15.zip') as inner_zip_file:
            # 내부 ZIP 파일을 다시 ZipFile 객체로 처리
            with zipfile.ZipFile(BytesIO(inner_zip_file.read())) as inner_zip:
                # CSV 파일을 메모리 상에서 바로 읽기
                for csv_name in inner_zip.namelist():
                    with inner_zip.open(csv_name) as csv_file:
                        df = pd.read_csv(csv_file)
                        df = df[df['dest_hdong_cd'] == code]
                        df_list.append(df)

    data = pd.concat(df_list, ignore_index=True)
    filename = f'od_population_{code}.csv'
    full_path = os.path.join('C:\\Users\\pjh08\\OneDrive\\바탕 화면\\', filename)
    data.to_csv(full_path, index=False)
    return data
```

```python
def od_population_group(data):
    age_under_10 = data[data['age'].isin([0,1])]['od_cnts'].sum()
    age_20_30 = data[data['age'].isin([2,3])]['od_cnts'].sum()
    age_40_50 = data[data['age'].isin([4,5])]['od_cnts'].sum()
    age_above_60 = data[data['age'].isin([6,7,8])]['od_cnts'].sum()

    daily_life = data[data['dest_purpose'].isin([0,1,2])]['od_cnts'].sum()
    leisure_life = data[data['dest_purpose'].isin([3,4,5])]['od_cnts'].sum()

    holiday_date = [20230902, 20230903, 20230909, 20230910, 20230916, 20230917, 20230923,
               20230924, 20230928, 20230929, 20230930, 20231001, 20231002, 20231003,
               20231007, 20231008, 20231009, 20231014, 20231015]

    weekday = data[~data['date'].isin(holiday_date)]['od_cnts'].sum()
    holiday = data[data['date'].isin(holiday_date)]['od_cnts'].sum()

    total_list = [age_under_10, age_20_30, age_40_50, age_above_60, daily_life, leisure_life,
                 weekday, holiday]

    return total_list
```

```python
group_list = []

for code in codes:
    group_list.append(od_population_group(od_population(code)))
    print(group_list)
```

```python
final_data = pd.DataFrame(group_list, columns = ['age_under_10', 'age_20_30', 'age_40_50', 'age_above_60',
          'daily_life', 'leisure_life', 'weekday', 'holiday'])
final_data
```
